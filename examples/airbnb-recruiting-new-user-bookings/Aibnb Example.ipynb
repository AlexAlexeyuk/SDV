### Airbnb Dataset Synthesis

This notebook shows a usage example over a sample of the Airbnb New User Bookings dataset
available on Kaggle: https://www.kaggle.com/c/airbnb-recruiting-new-user-bookings

Before running this notebook, make sure to having downloaded the data and run the
`Airbnb Dataset Sampling` notebook which can be found in the same folder as this one.

```python
from sdv import SDV

sdv = SDV()
sdv.fit('metadata_sample.json')
```

```python
real = sdv.metadata.get_tables()
```

```python
real['users'].sample(5)
```

```python
real['sessions'].sample(5)
```

```python
length = len(real['users'])
```

```python
length
```

```python
samples = sdv.sample_all(length, reset_primary_keys=True)
```

```python
samples['users'].sample(5)
```
